# 2. Gepaarter t-Test  / Differenztest

**Beispiel**: Der Datensatz survey aus dem Paket MASS enthält die Antworten von 237 Studierenden auf verschiedene Fragen. Unter anderem wurde die Spannweite der Schreibhand Wr.hnd und die Spannweite der Nicht-Schreibhand NW.hnd in cm abgefragt. Gibt es einen bedeutsamen Unterschied in der Spannweite?

```r
wd <- getwd()
wd
setwd(wd)
survey <- read.csv("survey.csv", header=TRUE)
str(survey)
```

```r
t.test(x = survey$Wr.Hnd, y = survey$NW.Hnd, paired = TRUE, conf.level = 0.95)
```

Anhand des p-Wertes ergibt sich also offenbar zum Niveau $\alpha = 0.05$ ein bedeutsamer Unterschied.

# 3.Nicht-gepaarter t-Test / Zweistichprobentest

Beispiel: Gibt es in der Spannweite der Schreibhand Wr.Hnd einen Unterschied im Hinblick auf das Geschlecht?

Zunächst ist hier eine grafische Veranschaulichung angebracht.

```r
geschlecht <- survey$Sex
boxplot(survey$Wr.Hnd ~ geschlecht)
```

```r
# bei Annahme gleicher Varianzen
t.test(survey$Wr.Hnd~geschlecht, paired = FALSE, var.equal = TRUE, conf.level = 0.95)

# bei Annahme ungleicher Varianzen
t.test(survey$Wr.Hnd~geschlecht, paired = FALSE, var.equal = FALSE, conf.level = 0.95)
```

Da das Konfidenzintervall nicht die Differenz von 0 enthält, gibt es basierend auf Signifikanzniveau von $\alpha = 0.05$ einen statistisch signifikanten Unterschied zwischen der Spannweite der Schreibhand im Hinblick auf das Geschlecht.
